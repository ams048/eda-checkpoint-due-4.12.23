## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a horizontal bar chart of the top N organizations by income amount

# Set the number of top organizations to consider
top_n = 10

# Sort the organizations by income amount (e.g., eo2 dataframe)
top_organizations = eo2.nlargest(top_n, 'INCOME_AMT')

# Set the figure size
plt.figure(figsize=(10, 6))

# Create the horizontal bar chart
plt.barh(top_organizations['NAME'], top_organizations['INCOME_AMT'])

# Set the title and labels
plt.title(f'Top {top_n} Organizations by Income Amount')
plt.xlabel('Income Amount')
plt.ylabel('Organization')

# Show the plot
plt.show()


```

```python

# Create a pie chart of the distribution of affiliations

# Extract the affiliation column from one of the dataframes (e.g., eo1)
affiliation_counts = eo1['AFFILIATION'].value_counts()

# Set the figure size
plt.figure(figsize=(8, 8))

# Create the pie chart
affiliation_counts.plot(kind='pie', autopct='%1.1f%%')

# Set the title and labels
plt.title('Distribution of Affiliations')
plt.ylabel('')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='STATE', data=df, order=df['STATE'].value_counts().head(10).index)
plt.title('Distribution of States')
plt.xlabel('State')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
